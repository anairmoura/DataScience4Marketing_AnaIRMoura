```python
#Compare Review Date with Visit month
reviews['reviewDate'] = pd.to_datetime(reviews['reviewDate'])
reviews['visitMonth'] = pd.to_datetime(reviews['visitMonth'])

#Calculate the difference between review date and visit month
reviews['review_lag_days'] = (reviews['reviewDate'] - reviews['visitMonth']).dt.days

#Filter out rows where review time is 0 or bellow 
reviews = reviews[reviews['review_lag_days'] >= 0]

print(reviews['review_lag_days'].head(15))
```
